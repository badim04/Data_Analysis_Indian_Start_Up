# Handling column "Amount"

```python
df["Amount($)"] = df["Amount($)"].replace("Undisclosed",0)
df["Amount($)"] = df["Amount($)"].replace("$Undisclosed",0)
df["Amount($)"] = df["Amount($)"].replace("Upsparks",0)
df["Amount($)"] = df["Amount($)"].replace("Series C",0)
df["Amount($)"] = df["Amount($)"].replace("Seed",0)
df["Amount($)"] = df["Amount($)"].replace("$undisclosed",0)
df["Amount($)"] = df["Amount($)"].replace("Pre-series A",0)
df["Amount($)"] = df["Amount($)"].replace("$",0)
```

```python
df["Amount($)"][677] = 0
df["Amount($)"][538] = 0
df["Amount($)"][551] = 0
```

```python
df["Amount($)"] = df["Amount($)"].fillna(0)
```

```python
df["Amount($)"].astype(float)
```

```python
df["Amount($)"] = df["Amount($)"].apply(lambda x : str(x).lstrip()) # removing spaces "$" sign
```

```python
df["Amount($)"] = df["Amount($)"].apply(lambda x : x[1:] if str(x).startswith("$") else x) # replacing "$" sign
```

```python
df["Amount($)"] = df["Amount($)"].apply(lambda x : str(x).replace(",","") ) # replacing "," sign
```

```python
df["Amount($)"] = df["Amount($)"].astype(float)
```
