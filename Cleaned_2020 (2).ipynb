```python
import pandas as pd
import numpy as np
```

```python
# Read the CSV file
df = pd.read_csv('startup_funding2020.csv')
```

```python
df.head()
```

```python
# Calculate mode of the "Founded" column then retrieve it and replace the blanks
founded_mode = df['Founded'].mode()[0]
df['Founded'].fillna(founded_mode, inplace=True)
```

```python
# Perform replacements in the 'Location' column
df['HeadQuarter'] = df['HeadQuarter'].replace(['Bangalore', 'Banglore'], 'Bangaluru', regex=True)

#Find the mode and replace the blanks with it
headquarter_mode = df['HeadQuarter'].mode()[0]
df['HeadQuarter'].fillna(headquarter_mode, inplace=True)

#Split HeadQuarter into city and country
df[['City', 'Country']] = df['HeadQuarter'].str.split(',', 1, expand=True)

#Replace Blanks in Country with India
df['Country'].fillna('India', inplace=True)
```

```python
df.head()
```

```python
#Performing Replacements
df['Sector'] = df['Sector'].replace(['Agriculture', 'Agritech', 'Agri tech'], 'AgriTech', regex=True)
df['Sector'] = df['Sector'].replace([' startup', ' Startup', ' tech', ' company', ' platform', ' Platform', ' Company'], '', regex=True)
df['Sector'] = df['Sector'].str.replace('Automobiles', 'Automobile', case=False)
df['Sector'] = df['Sector'].str.replace('Coworking', 'Co-working', case=False)

#Replace Blanks with None
df['Sector'].fillna('Undisclosed', inplace=True)
```

```python
#Replace Blanks with None
df['Founders'].fillna('Undisclosed', inplace=True)
df['Investor'].fillna('Undisclosed', inplace=True)
```

```python
# Perform replacement in the 'Amount($)' column
#df['Amount($)'] = df['Amount($)'].str.replace(',', '')
df['Amount($)'] = df['Amount($)'].replace(',', '', regex=True)
#df['Amount($)'] = df['Amount($)'].str.replace('$', '')
df['Amount($)'] = df['Amount($)'].str.replace('$', '', regex=True)

#Find the mean and replace the blanks with it
df['Amount($)'] = df['Amount($)'].replace(['Undisclosed', 'Undiclsosed', 'Undislosed'], '0', regex=True)

#Find the mean and replace the blanks with it
amount_mean = df['Amount($)'].mean()
df['Amount($)'].fillna(amount_mean, inplace=True)

# Convert 'Amount($)' column to integer data type
df['Amount($)'] = df['Amount($)'].astype(int)
```

```python
#Performing Replacements
df['Stage'] = df['Stage'].replace(['Pre seed', 'Pre- seed', 'Pre-seed'], 'Pre-Seed', regex=True)
df['Stage'] = df['Stage'].replace(['Preseries', 'Pre- series', 'Pre-series'], 'Pre-Series', regex=True)
df['Stage'] = df['Stage'].replace([' round', ' Round'], '', regex=True)
df['Stage'] = df['Stage'].replace([' Funding', ' Investment'], '', regex=True)

# Calculate mode of the "Stage" column then retrieve it and replace the blanks
stage_mode = df['Stage'].mode()[0]
df['Stage'].fillna(stage_mode, inplace=True)
```

```python
df.head()
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv('2020_cleaned_dataset.csv', index=False)
```
